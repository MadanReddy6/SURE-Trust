```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./cause_of_deaths.csv")
df
```

```python
x = df[df["Country/Territory"] == "India"]
x = x.sort_values("Year")
```

```python
plt.plot(x.Year, x.Malaria)
plt.show()
```

```python
plt.bar(x.Year, x.Malaria)
plt.show()
```

```python
x = []
y = []
for country, c_df in df.groupby("Country/Territory"):
    mal_death = c_df["Malaria"].sum()
    if mal_death > 50000:
        x.append(country)
        y.append(mal_death)
```

```python
plt.figure(figsize=(10,10))
plt.xlabel("Number of deaths (in millions)")
plt.barh(x,y)
plt.plot()
```

```python

x , y = [], []
for country, c_df in df.groupby("Country/Territory"):
    death = c_df["Nutritional Deficiencies"].mean()
    x.append(country)
    y.append(death)
```

```python
data = {
    "countries" : x,
    "deaths" : y
}

data = pd.DataFrame(data)
top_30 = data.sort_values("deaths").tail(30)
```
