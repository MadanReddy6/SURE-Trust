```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
adm_df = pd.read_csv("./Admission_Predict_.csv", index_col="Serial No.")
adm_df.info()

adm_df["Chance of Admit"] = adm_df["Chance of Admit"]*100
```

```python
# step 1 
# calculate the loss (keep the formula in mind)
    # - make a prediction


# step 2
# calculate the gradient, based on the formula

# step 3 
# if gradient is  +ve, we decrease the slope, if grad is -ve , we increase the slope

# step 4
# keep repeating


weight = 1
bias = 1
```

```python
plt.scatter(y = adm_df["Chance of Admit"], x= adm_df["CGPA"])
plt.ylabel("Probablity")
plt.xlabel("CGPA")

adm_df["Prediction"] = weight*adm_df["CGPA"] + bias

plt.plot(adm_df["CGPA"], adm_df["Prediction"], c="Red")
plt.show()
```

```python
def mse(y , y_):
    return sum((y-y_)**2)/len(y)

def grad(y, y_, x):
    return sum(-2*(y - y_)*x)/len(y) , sum(-2*(y - y_))/len(y)

def pred(w, b, x):
    return w*x + b
```

```python
weight = 0
bias = 0
lr = 0.01
```

```python
for i in range(10000):
    
    feature = adm_df["CGPA"]
    label = adm_df["Chance of Admit"]
    
    prediction = pred(weight, bias, feature)
    loss = mse(label , prediction)
    
    w_grad , b_grad = grad(label, prediction, feature)
    
    weight = weight - w_grad*lr
    bias = bias - b_grad*lr
    
print(loss , weight, bias)
```

```python
cgpa = 5
out = (weight * cgpa) + bias

print(out)
```
