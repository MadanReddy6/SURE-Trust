```python
import matplotlib.pyplot as plt
```

```python
x = [1,34,2,53,52,34,32,34,56,24,34]

# fig, ax = plt.subplots(2,3)

plt.subplot(2,2, 4)
plt.plot(x,x)

plt.subplot(2,2, 3)
plt.scatter(x,x)
plt.xlabel("X")

plt.subplot(2,2, 2)
plt.bar(x,x)

plt.subplot(2,2, 1)
plt.title("Pie")
plt.pie(x)

figure = plt.figure(figsize=(20,40))
# figure.subplots_adjust(hspace=1, wspace=1)
plt.show()


```
